```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "data-1607396405939.csv"

merged_data = pd.read_csv(path)
```

```python
merged_data.head()
```

```python
merged_data['date'] = pd.to_datetime(merged_data['date'], format='%Y-%m-%d')
merged_data
```

```python
cases = merged_data['total_cases']
new_cases = merged_data['new_cases']
parks = merged_data['parks_percent_change_from_baseline']
dates = merged_data['date']

plt.scatter(cases, parks, marker="o")
plt.title(f"Total cases vs. Parks Percent Change")
plt.xlabel("Total Cases")
plt.ylabel("% Change from Baseline - Parks")
plt.show()
```

```python
plt.plot(dates, cases, color="red", label="Cases")
plt.xticks(rotation=45)
plt.title("Indiana - Total Cases")
plt.savefig("total_cases.png")
plt.show()
```

```python

plt.plot(dates, parks, color="blue", label="Cases")
plt.xticks(rotation=45)
plt.title("Indiana - Parks Mobility")
plt.savefig("mobility.png")
plt.show()
```

```python
plt.plot(dates, new_cases, color="blue", label="Cases")
plt.xticks(rotation=45)
plt.title("Indiana - New Cases")
plt.savefig("new_cases.png")
plt.show()
```
