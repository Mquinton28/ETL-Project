```python
import pandas as pd
import os
from sqlalchemy import create_engine
import datetime
```

```python
df = pd.read_csv("2020_US_Region_Mobility_Report.csv")
df
```

```python
clean_df = df.rename(columns={"country_region_code": "country_code",
                             "country_region": "country",
                             "sub_region_1": "state",
                             "sub_region_2": "county",
                             "metro_area": "metro_area",
                             "iso_3166_2_code": "territory_subdivisions",
                             "census_fips_code": "census_fips_code",
                             "date": "date",
                             "retail_and_recreation_percent_change_from_baseline": "retail_and_recreation_percent_change_from_baseline",
                             "grocery_and_pharmacy_percent_change_from_baseline": "grocery_and_pharmacy_percent_change_from_baseline",
                             "parks_percent_change_from_baseline": "parks_percent_change_from_baseline",
                             "transit_stations_percent_change_from_baseline": "transit_stations_percent_change_from_baseline",
                             "workplaces_percent_change_from_baseline": "workplaces_percent_change_from_baseline",
                             "residential_percent_change_from_baseline": "residential_percent_change_from_baseline",})
clean_df.head()
```

```python
clean_df = clean_df[[
    "state",
    "county",
    "date",
    "retail_and_recreation_percent_change_from_baseline",
    "grocery_and_pharmacy_percent_change_from_baseline",
    "parks_percent_change_from_baseline",
    "transit_stations_percent_change_from_baseline",
    "workplaces_percent_change_from_baseline",
    "residential_percent_change_from_baseline"
]]
clean_df.head()
```

```python
clean_df = clean_df.dropna(subset=['state'])
clean_df
```

```python
indiana_df = clean_df[clean_df['state'] == "Indiana"]
indiana_df
```

```python
indiana_df.dtypes
```

```python
indiana_df['date'] = pd.to_datetime(indiana_df['date'], format='%Y%m%d', infer_datetime_format=True)
```

```python
indiana_df.dtypes
```

```python
indiana_df_drop_na = indiana_df.dropna(subset=['county'])
```

```python
indiana_df_reset = indiana_df_drop_na.reset_index(drop=True)
indiana_df_reset
```

```python
indiana_df2 = indiana_df_reset.set_index(['date'])
indiana_df2
```

```python
indiana_county_mean = indiana_df2.groupby(['county']).mean()
indiana_county_mean
```

```python
indiana_df2_mean = indiana_df2.groupby(['date']).mean()
indiana_df2_mean
```

## Create Database Connection

```python
conn = 'postgres:postgres@localhost:5432/covid-indiana'
engine = create_engine(f'postgresql://{conn}')
```

```python
engine.table_names()
```

```python
indiana_df2_mean.to_sql(name='mobility_per_date', con=engine, if_exists='append', index=True)
```
